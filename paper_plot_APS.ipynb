```python
# Function for the eigenvalues of the Hamiltonian.
def construct_eigenvalues(L, k):
     eigvals_lst = []
     for j in range(L-k+1+1):
          lambda_L_j = 3**(k/2) * (2*j - (L-k+1)) 
          degeneracy = 2**(k-1) * math.comb(L-k+1 , j)
          for _ in range(degeneracy):
               eigvals_lst.append(lambda_L_j)
     return np.array(eigvals_lst)

def count_adjacent_k_local_pauli_strings(L, k):
    return (L - k + 1) * (3 ** k)

# path = '/Users/sasankadowarah/python_test_codes/'+str(folder_name)+'/'
path = '/Volumes/SASANKA_SSD/quantum_chaos_data/long_range_hamiltonians/'+str(folder_name)+'/'
os.chdir(path)
b = 0
os.chdir(path + "b" + str(b)) 

L = 9
# number_of_pauli_strings = np.load('2_local_paulis_length.npy')
# print('Number of Pauli strings generated:', number_of_pauli_strings)
# print('NUmber of adjacent k-local Pauli strings:', count_adjacent_k_local_pauli_strings(L, 2))

iter_lst = [i for i in range(0, 160)]

mu = np.load('mu.npy')
sigma = np.load('sigma.npy')

print('mu:', mu)
print('sigma:', sigma)
```

```python
import numpy as np
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt

def unfold_spectrum(L, k, eigenvalues, degree=7, plot=False):
    """
    Unfolds a spectrum by fitting the integrated density of states (IDOS) 
    with a polynomial and applying the mapping E → ε = N(E).
    
    Parameters:
        eigenvalues (1D array): Sorted raw eigenvalues.
        degree (int): Degree of polynomial fit for smooth IDOS.
        plot (bool): If True, show plots of IDOS and spacing histograms.
    
    Returns:
        unfolded (1D array): Unfolded eigenvalues.
        spacings (1D array): Level spacings of unfolded spectrum.
    """
    # Handle degeneracy: for odd L and even k, select every other eigenvalue
    if (L % 2 == 1) and (k % 2 == 0):
        eigenvalues = eigenvalues[::2]

    # Sort and compute IDOS
    E_sorted = np.sort(eigenvalues)
    N = np.arange(1, len(E_sorted)+1)

    # Fit polynomial to IDOS
    p = Polynomial.fit(E_sorted, N, degree)
    N_smooth = p(E_sorted)

    # Unfolded eigenvalues = smooth IDOS
    unfolded = N_smooth
    spacings = np.diff(unfolded)

    if plot:
        # Plot IDOS
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(E_sorted, N, 'b.', label='Staircase N(E)')
        plt.plot(E_sorted, N_smooth, 'r-', label=f'Poly fit (deg={degree})')
        plt.xlabel("E"); plt.ylabel("N(E)")
        plt.title("Integrated Density of States")
        plt.legend()

        # Plot spacing histogram
        plt.subplot(1, 2, 2)
        plt.hist(spacings / np.mean(spacings), bins=50, density=True, alpha=0.7, color='burlywood')
        plt.xlabel("s (rescaled)")
        plt.ylabel("P(s)")
        plt.title("Spacing Distribution After Unfolding")
        plt.tight_layout()
        plt.show()

    return unfolded, spacings
```
